# CORRELACIONES ANALISIS

## 📊 Observaciones Clave en el Rendimiento Académico de Estudiantes

---

### 1. 🎓 Relación entre Calificaciones (G1, G2, G3)
- Las variables **G1**, **G2**, y **G3** (calificaciones en diferentes momentos) muestran correlaciones altas entre sí (cercanas a 1). 📈
- **Interpretación**: Los estudiantes que obtienen buenas calificaciones en una evaluación tienden a mantener ese rendimiento en otras, lo cual refleja un rendimiento académico consistente. ✅

---

### 2. 👩‍🏫 Correlación de Calificaciones con Educación de los Padres (Medu y Fedu)
- **Medu** (educación de la madre) tiene una correlación positiva moderada con **G1**, **G2**, y **G3**.
- **Fedu** (educación del padre) también tiene una correlación positiva, aunque un poco más débil.
- **Interpretación**: Los estudiantes cuyos padres tienen un nivel educativo más alto tienden a obtener mejores calificaciones. Esto sugiere que el nivel educativo de los padres influye positivamente, probablemente por el ambiente de apoyo en casa. 🏡

---

### 3. 📚 Efecto del Tiempo de Estudio y Ausencias en Calificaciones
- **Studytime** (tiempo de estudio) muestra una correlación positiva con las calificaciones, aunque no es particularmente fuerte.
- **Absences** (ausencias) tiene una correlación negativa, especialmente con **G3**.
- **Interpretación**: Los estudiantes que asisten regularmente tienden a obtener mejores calificaciones. Esto refuerza la importancia de la asistencia en el rendimiento académico. ✅

---

### 4. 🍻 Consumo de Alcohol y Relaciones Sociales
- **Dalc** (consumo diario de alcohol) y **Walc** (consumo de alcohol los fines de semana) tienen correlaciones negativas débiles con las calificaciones, especialmente con **G3**.
- **Goout** (frecuencia de salir) también muestra una correlación negativa.
- **Interpretación**: El consumo de alcohol, especialmente durante los fines de semana, y una vida social activa pueden afectar negativamente el rendimiento académico, aunque no son los factores más influyentes. 🛑

---

### 5. 🏠 Relaciones con el Tipo de Familia y Apoyo Familiar
- **Famsize_LE3** (familia de 3 o menos miembros) y **Pstatus_T** (padres viviendo juntos) no muestran correlaciones fuertes con las calificaciones.
- **Famsup_yes** (apoyo educativo familiar) tiene una correlación positiva débil con las calificaciones.
- **Interpretación**: Aunque el tipo de familia y el tamaño no son determinantes, el apoyo educativo familiar tiene una influencia positiva moderada en el rendimiento académico. 👨‍👩‍👦

---

### 6. 🎓 Variables Sociales y de Apoyo Escolar
- **Activities_yes** (participación en actividades extracurriculares) y **internet_yes** (acceso a internet en casa) no muestran correlaciones fuertes con las calificaciones.
- **Higher_yes** (interés en estudios superiores) tiene una correlación positiva débil con las calificaciones.
- **Interpretación**: Los estudiantes interesados en continuar con estudios superiores tienden a obtener un rendimiento académico ligeramente mejor. 📈

---

## 📌 Resumen General
- **Factores que muestran correlaciones positivas con las calificaciones**:
  - 🎓 Nivel educativo de los padres
  - 📚 Tiempo de estudio y asistencia regular
  - 👨‍👩‍👦 Apoyo familiar
  - 🎓 Interés en estudios superiores

- **Factores que muestran correlaciones negativas con las calificaciones**:
  - 🍻 Ausencias, consumo de alcohol, y frecuencia de salir con amigos

## 📝 Conclusión
En términos generales, los factores que parecen tener una mayor influencia en el rendimiento académico de los estudiantes en este conjunto de datos son:

- **Educación de los padres**: Esto indica que el apoyo y el contexto educativo en casa pueden ser importantes. 🏠
- **Asistencia y tiempo de estudio**: La regularidad en asistir a clases y dedicar tiempo al estudio tienen una influencia positiva en las calificaciones. ✅
- **Consumo de alcohol y vida social**: Aunque no tienen una influencia extremadamente fuerte, el consumo de alcohol y la vida social parecen afectar negativamente el rendimiento académico. 🍻🛑
