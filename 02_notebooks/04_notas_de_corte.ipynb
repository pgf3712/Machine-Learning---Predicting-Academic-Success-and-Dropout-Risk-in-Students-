```python
def clasificar_titulacion(titulacion):
    biosanitario = ["Medicina", "Enfermería", "Psicología", "Farmacia", "Fisioterapia", "Nutrición", "Veterinaria", "Odontología", "Terapia Ocupacional", "Logopedia", "Podología", "Neurociencia", "Óptica", "Optometría"]
    ciencias = ["Biología", "Química", "Física", "Matemáticas", "Geología", "Ciencias Ambientales", "Estadística", "Bioquímica", "Ciencia", "Nanociencia", "Astronomía", "Biotecnología", "Geografía", "Recursos Hídricos"]
    tecnologico_ingenierias = ["Ingeniería", "Tecnología", "Computadores", "Telecomunicación", "Informática", "Electrónica", "Aeroespacial", "Materiales", "Robótica", "Videojuegos", "Desarrollo", "Industrial", "Arquitectura", "Edificación", "Sistemas de Información", "PARS", "Sistemas Informáticos Distribuidos", "Inteligencia Artificial", "Análisis de Datos", "Computación"]
    ciencias_sociales_juridicas = ["Derecho", "Economía", "Administración", "Relaciones Internacionales", "Sociología", "Ciencias Políticas", "Antropología", "Criminología", "Trabajo Social", "Educación", "Pedagogía", "Finanzas", "Contabilidad", "ADE", "Marketing", "Protocolo", "Comercio", "Periodismo", "Publicidad", "Relaciones Laborales", "Turismo", "Información y Documentación", "Gestión de la Información", "Dirección y Gestión", "Gestión Aeronáutica", "Gestión y Operaciones del Transporte Aéreo", "European Studies"]
    humanidades_artes = ["Historia", "Arte", "Filosofía", "Lingüística", "Literatura", "Diseño", "Bellas Artes", "Arqueología", "Estudios", "Traducción", "Música", "Teatro", "Cine", "Danza", "Humanidades", "Estudios Clásicos", "Estudios Hispánicos", "Filología", "Conservación", "Patrimonio", "Musicología", "Composición", "Audiovisual", "Fotografía", "Lenguas Modernas", "Paisajismo", "Comunicación Digital"]
    
    if "Maestro de Ed. Infantil" in titulacion or "Maestro de Educación Primaria" in titulacion:
        return "Ciencias Sociales-Jurídicas"
    if any(keyword in titulacion for keyword in biosanitario):
        return "Biosanitario"
    elif any(keyword in titulacion for keyword in ciencias):
        return "Ciencias"
    elif any(keyword in titulacion for keyword in tecnologico_ingenierias):
        return "Tecnológico-Ingenierías"
    elif any(keyword in titulacion for keyword in ciencias_sociales_juridicas):
        return "Ciencias Sociales-Jurídicas"
    elif any(keyword in titulacion for keyword in humanidades_artes):
        return "Humanidades y Artes"
    else:
        return "Sin Clasificar"


df_notas_corte_total['Especialidad'] = df_notas_corte_total['TITULACIONES DE GRADO'].apply(clasificar_titulacion)
df_notas_corte_total.head()
```
